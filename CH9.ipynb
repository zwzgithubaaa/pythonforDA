### Figure 和 Subplot
* matplotlib的图像都位于Figure对象中：用plt.figure创建新的Figure

```python
#创建新的Figure
fig = plt.figure()
#不能通过空的Figure绘图，必须用add_subplot创建一个或多个subplot
ax1 = fig.add_subplot(2,2,1)#图像是2*2的（最多4张），当前选中的是4个subplot中的第一个
ax2 = fig.add_subplot(2,2,2)#再把后面两个subplot创建出来
#jupyternotebook在每个小窗之后图形被重置，必须将所有的绘图命令存在一个小窗里。
plt.plot(np.random.randn(50).cumsum(),'k--')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot([1.5,3.5,-2,1.6])#在随后一个用过的subplot上进行绘制
plt.plot(np.random.randn(50).cumsum(),'k--')#'k--'是一个线性选项，黑色虚线
```

```python
#fig.add_subplot返回AxesSubplot对象，直接调用即在其空着的格子画图
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
```
